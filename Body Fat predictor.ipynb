```python
def distplt(col):
    plt.figure(figsize=(12,5))
    sns.distplot(data['BodyFat'],color='magenta',hist=False,label='BodyFat')
    sns.distplot(data[col],color='red',hist=False,label=col)
    plt.legend()
    plt.show()
cols = list(data.columns) 
for i in cols:
    print(f'Distribution plots for {i} feature is shown below')
    distplt(i)
    print("-"*100)
```

```python
import scipy.stats as stats
```

```python
def drawplots(data,col):
    plt.figure(figsize=(15,7))
    
    plt.subplot(1,3,1)
    plt.hist(data[col],color='crimson')
    
    plt.subplot(1,3,2)
    stats.probplot(data[col],dist='norm',plot=plt)
    
    plt.subplot(1,3,3)
    sns.boxplot(data[col],color='magenta')
    
    plt.show()
    
cols = list(data.columns)
for i in range(len(cols)):
    print(f'Distribution plots for the featire {cols[i]} are shown below')
    drawplots(data,cols[i])
    print("="*100)
```

```python
#Checking for outlaiers
upper_limit = []
lower_limit = []
for i in data.columns:
    upper_limit.append(data[i].mean() + (data[i].std())*4)
    lower_limit.append(data[i].mean() - (data[i].std())*4)
```

```python
cols = list(data.columns)
j=0
for i in range(len(cols)):
    temp = data.loc[(data[cols[i]]>upper_limit[j])&(data[cols[i]]<lower_limit[j])]
    j+=1
```

```python
def correlation(data,threshold):
    colcor = set()
    cormat = data.corr()
    for i in range(len(cormat)):
        for j in range(i):
            if abs(cormat.iloc[i][j]) > threshold:
                colname = cormat.columns[i]
                colcor.add(colname)
    return colcor
ans = correlation(train,threshold=0.85)
ans
```

```python
totrain = train[cols1]
totrain.head()
```

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV
from sklearn import metrics
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression,Lasso,Ridge
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(totrain,test,test_size=0.2)
```

```python
model_DTR = DecisionTreeRegressor(ccp_alpha=1)
model_DTR.fit(xtrain,ytrain)
```

```python
ypred = model_DTR.predict(xtest)
print(f'Decision Tree Noraml aproach:{metrics.r2_score(ytest,ypred)}')
```

```python
model_RFR = RandomForestRegressor(n_estimators=1000,ccp_alpha=1)
model_RFR.fit(xtrain,ytrain)
```

```python
ypred_rfr = model_RFR.predict(xtest)
print(f'Random Forest Normal approach:{metrics.r2_score(ytest,ypred_rfr)}')
```

```python
params = {
  'RandomForest':{  
    'model':RandomForestRegressor(),
    'params':{
        'n_estimators':[int(x) for x in np.linspace(start=1,stop=1200,num=10)],
        'criterion':["mse","mae"],
        'max_depth':[int(x) for x in np.linspace(start=1,stop=30,num=5)],
        'min_samples_split':[2,5,10,12],
        'min_samples_leaf':[2,5,10,12],
        'max_features':['auto','sqrt'],
        'ccp_alpha':[1,2,2.5,3,3.5,4,5],
    }
 },
  'D-tree': {
    'model':DecisionTreeRegressor(),
    'params':{
         'criterion':["mse","mae"],
         'splitter':['best','random'],
         'min_samples_split':[1,2,5,10,12],
         'min_samples_leaf':[1,2,5,10,12],
         'max_features':['auto','sqrt'],
         'ccp_alpha':[1,2,2.5,3,3.5,4,5],
     }
  },
  'SVM':{
    'model':SVR(),
    'params':{
          'C':[0.25,0.50,0.75,1.0],
          'tol':[1e-10,1e-5,1e-4,0.025,0.50,0.75],
          'kernel':['linear','poly','rbf','sigmoid'],
          'max_iter':[int(x) for x in np.linspace(start=1,stop=250,num=10)],
    }
  }
}   
 
```

```python
scores = []
for modelname,mp in params.items():
    clf = RandomizedSearchCV(mp['model'],param_distributions=mp['params'],cv=5,n_jobs=-1,n_iter=10,scoring='neg_mean_squared_error')
    clf.fit(xtrain,ytrain)
    scores.append({'model_name':modelname,'best_score':clf.best_score_,'best_estimator':clf.best_estimator_})
```

```python
score_df = pd.DataFrame(scores,columns=['model_name','best_score','best_estimator'])
score_df
```

```python
scores[0]['best_estimator']
```

```python
rf = scores[0]['best_estimator']
rf.fit(xtrain,ytrain)
```

```python
ypred = rf.predict(xtest)
print(metrics.r2_score(ytest,ypred))
```
